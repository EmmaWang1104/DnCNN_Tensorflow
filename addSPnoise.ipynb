resize resert train and test images to 180x180 and add salt and pepper noise

```python
import os
import numpy as np
import cv2
import glob
```

```python
src_dir = '/content/drive/MyDrive/coen344-cv1-project/data/train/'
save_dir = '/content/drive/MyDrive/coen344-cv1-project/data/train/SP/'
src_dir_test = '/content/drive/MyDrive/coen344-cv1-project/data/test/'
save_dir_test = '/content/drive/MyDrive/coen344-cv1-project/data/test/SP/'

filepaths = glob.glob(src_dir + '*.jpg')
filepaths_test = glob.glob(src_dir_test + '*.jpg')

def sortKeyFunc(s):
    return int(os.path.basename(s)[:-4])

filepaths.sort(key=sortKeyFunc)
filepaths_test.sort(key=sortKeyFunc)

print("[*] Reading train files...")  

if not os.path.exists(os.path.join(save_dir, "original-sp")):
    os.makedirs(os.path.join(save_dir, "original-sp"))
if not os.path.exists(os.path.join(save_dir, "noisy-sp")):
    os.makedirs(os.path.join(save_dir, "noisy-sp"))
if not os.path.exists(os.path.join(save_dir_test, "original-sp")):
    os.makedirs(os.path.join(save_dir_test, "original-sp"))
if not os.path.exists(os.path.join(save_dir_test, "noisy-sp")):
    os.makedirs(os.path.join(save_dir_test, "noisy-sp"))
```

```python
def getSPNoisy(img): 
  size,_,_ = img.shape
  
  s_vs_p = 0.5
  amount = 0.01
  noisy = np.copy(img)

  num_salt = np.ceil(amount * img.size * s_vs_p) #number of pixels to turn white
  num_pepper = np.ceil(amount* img.size * (1. - s_vs_p)) #number of pixels to turn black
  r = np.random.permutation(size*size)

  i = 0
  while i < num_salt:
    m = r[i]//size
    n = r[i]%size
    noisy[m,n,0] = 255
    noisy[m,n,1] = 255
    noisy[m,n,2] = 255
    i+=1
  while i < num_salt+num_pepper:
    m = r[i]//size
    n = r[i]%size
    noisy[m,n,0] = 0
    noisy[m,n,1] = 0
    noisy[m,n,2] = 0
    i+=1

  return noisy.astype('uint8')
```

```python
def process(filepaths, save_dir):
  for i, filepath in enumerate(filepaths):
    image = cv2.imread(filepath)
    image = cv2.resize(image, (180, 180), interpolation=cv2.INTER_CUBIC)
    noisy = getSPNoisy(image)

    cv2.imwrite(os.path.join(save_dir, "noisy-sp", f"{i:04d}.png"), noisy)
    cv2.imwrite(os.path.join(save_dir, "original-sp", f"{i:04d}.png"), image)
```

```python
print("[*] Applying noise...")

process(filepaths,save_dir)
process(filepaths_test,save_dir_test)

print("[*] Noisy and original images saved")
```
